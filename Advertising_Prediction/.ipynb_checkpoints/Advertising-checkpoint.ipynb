### EDA Bivariate Analysis

```python
#work with oridnal
dic={
    'A':0,
    'B':1,
    'C':2,
    'D':3,    
}
df_c['class']=df_c['class'].map(dic).astype(int)
df_c
```

```python
#work with nominal
#one hot encoder
df_c['influencer'].nunique()
df_c['influencer'].unique()
df_c['influencer'].value_counts()
```

```python
lst_small=['influencer']
#create one hot encoder object
ohe = OneHotEncoder(sparse=False )
df_ohe = pd.DataFrame(ohe.fit_transform(df_c[lst_small]) , columns=ohe.get_feature_names_out())
df_ohe
```

### Modeling 

```python
df_c = pd.concat([df_c , df_ohe ] , axis = 1)
df_c.drop(lst_small , axis = 1 , inplace = True )
```

```python
df_c
```

```python
df_c.dropna(inplace=True)
```

```python
df_c.isna().sum()
```

```python
df_c.reset_index(inplace=True)
df_c
```

```python
df_c.drop('index',axis=1,inplace=True)
df_c
```

```python
#split data
x , y = df_c.drop(['sales'] , axis = 1 ) , df_c['sales']
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state = 0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#work with numerical data
#create standard scaler object
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
#modeling and evaluation
#model 1  --- > linear regression
#create linear regression object
l_reg = LinearRegression()
l_reg.fit(x_train_scaled, y_train)

print("Training set score: ",np.around(l_reg.score(x_train_scaled, y_train),2)* 100,'%' )
print("Test set score: ",np.around(l_reg.score(x_test_scaled, y_test),2)* 100,'%')
```

```python
#predict and show error rate
from sklearn.metrics import r2_score , mean_squared_error
y_train_pred = l_reg.predict(x_train_scaled)
y_test_pred = l_reg.predict(x_test_scaled)
print("Training set pred score: ",np.around(r2_score(y_true = y_train,y_pred = y_train_pred),2)* 100,'%' )
print("Test set  pred score: ",np.around(r2_score(y_true = y_test,y_pred = y_test_pred),2)* 100,'%')
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_test_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_test_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
```

```python
#coeffs
print(l_reg.intercept_)
print(l_reg.coef_)
```

```python
#model 2  --- > Poly regression
#create poly regression object
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=3)
x_train = poly.fit_transform(x_train_scaled)
x_test = poly.transform(x_test_scaled)
sclr = StandardScaler()
x_train_scaled = sclr.fit_transform(x_train)
x_test_scaled = sclr.transform(x_test)
l_reg = LinearRegression()
l_reg.fit(x_train_scaled, y_train)
print("Training set score: ",np.around(l_reg.score(x_train_scaled, y_train),2)* 100,'%' )
print("Test set score: ",np.around(l_reg.score(x_test_scaled, y_test),2)* 100,'%')
```

```python
#predict and show error rate
from sklearn.metrics import r2_score , mean_squared_error
y_train_pred = l_reg.predict(x_train_scaled)
y_test_pred = l_reg.predict(x_test_scaled)
print("Training set pred score: ",np.around(r2_score(y_true = y_train,y_pred = y_train_pred),2)* 100,'%' )
print("Test set  pred score: ",np.around(r2_score(y_true = y_test,y_pred = y_test_pred),2)* 100,'%')
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_test_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_test_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
```

```python
#coeffs
print(l_reg.intercept_)
print(l_reg.coef_)
```
